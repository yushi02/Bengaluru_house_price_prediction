```python
df=pd.read_csv("/kaggle/input/bengaluru-house-price-data/Bengaluru_House_Data.csv")
df
```

```python
df.groupby('area_type').agg('count')
```

```python
df=df.drop(['area_type','availability','society','balcony'],axis='columns')
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
df
```

```python
df['size'].unique()
```

```python
df['bhk'] = df['size'].apply(lambda x: int(x.split(' ')[0]))
df.bhk.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
df.total_sqft.unique()
df[~df['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_to_sqft(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        float(x)
    except:
        return None
```

```python
df1=df.copy()
df1.total_sqft=df1.total_sqft.apply(convert_to_sqft)
df1.head()
df1=df1.dropna()
df1.head(2)
```

```python
df2=df1.copy()
df2['price_per_sqft']=df2['price']*100000/df2['total_sqft']
df2
```

```python
df2_stats=df2['price_per_sqft'].describe()
df2_stats
```

```python
df2.to_csv("bhp.csv",index=False)
df2.shape
```

```python
df2.location=df2.location.apply(lambda x: x.strip())
location_stats=df2['location'].value_counts(ascending=False)
location_stats
```

```python
len(location_stats[location_stats>2])
location_stats_less_than_2 = location_stats[location_stats<=2]
```

```python
df2.location=df2.location.apply(lambda x: 'other' if x in location_stats_less_than_2 else x)
df2
```

```python
df2[~(df2.total_sqft/df1.bhk<300)].describe()
```

```python
X=df2.drop(['price','location','size'],axis='columns')
y=df2.price
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
